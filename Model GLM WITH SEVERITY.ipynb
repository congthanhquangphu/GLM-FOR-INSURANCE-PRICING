# Creating and Training the Model

```python
import statsmodels.api as sm
import numpy as np
import statsmodels.formula.api as smf
```

```python
fomular = 'CLAIM ~  NB_SEATS + USE + SUM_INSURED_Group + AGE_INTERVALS + CAPACITY_GROUP + CAR_MAKER '
```

```python
Gamma_model = smf.glm(formula=fomular,data=train, family=sm.families.Gamma(sm.families.links.log))
```

```python
Gamma_results = Gamma_model.fit()
```

```python
Gamma_results.summary()
```

```python
Gamma_results.aic
```

```python
b = Gamma_results.predict(test_x)
```

```python
test['CLAIM_Predict'] = b
```

```python
test['|Y-Y^|'] = test['CLAIM'] - test['CLAIM_Predict']
```

```python
test['|Y-Y^|']=test['|Y-Y^|'].abs()
```
