# Creating and Training the Model

```python
import statsmodels.api as sm
import numpy as np

from sklearn.model_selection import train_test_split
import statsmodels.formula.api as smf
```

```python
Binomial_model = sm.GLM(train_y,train_x, family=sm.families.NegativeBinomial())
```

```python
Binomial_results = Binomial_model.fit()
```

```python
print(Binomial_results.summary())
```

```python
Binomial_results.bic
```

```python
Binomial_results.aic
```

```python
b = Binomial_results.predict(test_x)
```

```python
test['NB.CLAIM_Predict'] = b
```

```python
test['|Y-Y^|'] = test['NB.CLAIM'] - test['NB.CLAIM_Predict']
```

```python
test['|Y-Y^|']=test['|Y-Y^|'].abs()
```
