```python
def split_year(select_years):
    result = pd.DataFrame(columns=select_years.columns)
    for index, row in select_years.iterrows():
        while row['Nb_years'] > 0:
            new_row = row.copy(deep=True)
            year = new_row['Start_seq'].year
            new_row['End_seq'] = pd.Timestamp(year=year, month=12, day=31)
            new_row['Nb_years'] = 0
            result = result.append(new_row)

            row['Start_seq'] = pd.Timestamp(year=year + 1, month=1, day=1)
            row['Year'] = row['Start_seq'].year
            row['Nb_years'] = row['End_seq'].year - row['Year']

        result = result.append(row)
    print(result[['CERTIFICATE','Start_seq', 'End_seq', 'Year', 'Nb_years']])
    return result
```

```python
""" Handle the 1-year-diff by split them in two"""
select_years1 = df2.loc[df2['Nb_years'] == 1]
select_years2 = select_years1.copy(deep=True)
print(select_years1[['Start_seq', 'End_seq', 'Year', 'Nb_years']])

""" First split"""
select_years1['End_seq'] = pd.to_datetime(
    (select_years1.Year*10000+12*100+31), format='%Y%m%d')
select_years1['Nb_years'] = 0

""" Second split"""
select_years2['Start_seq'] = pd.to_datetime(
    ((select_years2.Year+1)*10000+1*100+1), format='%Y%m%d')
select_years2['Year'] = select_years2.Year+1
select_years2['Nb_years'] = 0
```

```python
""" Print"""
diff1_year = pd.concat([select_years1, select_years2])
print(diff1_year[['Start_seq', 'End_seq', 'Year', 'Nb_years']])
```
